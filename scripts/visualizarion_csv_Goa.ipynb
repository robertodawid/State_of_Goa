## OPERATING COST BY SECTOR

```python
ope_cost = n_file_r[n_file_r['NAME']=='DiscOperatingCost']
ope_cost.drop(ope_cost.iloc[:, 2:13], axis=1, inplace=True)

#power tech
ope_cost = ope_cost.replace({'TECHNOLOGY': di2})

#power tech RE imports
ope_cost = ope_cost.replace({'TECHNOLOGY': di3})

#Transport
di5 = ['FTR', 'PTR', 'TRAN', 'MINATF']
ope_cost.loc[ope_cost['TECHNOLOGY'].str.contains('|'.join(di5)),'TECHNOLOGY'] = 'Transport'

#othres
di6 = ['PWRTRN', 'WHSOL', 'HWT', 'STO', 'APP'] #others, to filter them from the main data frame, usually these have the residual
ope_cost.loc[ope_cost['TECHNOLOGY'].str.contains('|'.join(di6)),'TECHNOLOGY'] = 'Other'

ope_cost.rename(columns={'TECHNOLOGY':'Sector'}, inplace=True)

ope_cost.head(5)
```

```python
# list of operating costs sorted by the dictionaries in the previous cell

calc_ope_cost_l = ['CCGT',
 'Coal PP',
 'IEX-Imports',
 'Import Coal PP',
 'Nuclear',
 'Other',
 'RE import Wind',
 'RE-Imports',
 'Small hydro',
 'Solar RPO',
 'Solar-PV Agriculture',
 'Transport',
 'Wind',
 'Wind RPO',
 'Solar-ground+float',
 'Solar-PV Residential',
 'Solar-PV Commercial',
 'Biomass',
 'RE import Solar']
```

```python
#Calculated
oper_cost_A = ope_cost.loc[ope_cost['Sector'].isin(calc_ope_cost_l)]

#no Calculated
oper_cost_B = ope_cost.loc[~ope_cost['Sector'].isin(calc_ope_cost_l)]

#dictionary to change the Sector names for the fuel names
di_f = {'IMPCOA':'COA', 'IMPDSL':'DSLS',
        'IMPKER':'KER', 'IMPLPG':'LPG',
        'IMPNGS': 'NGSS', 'MINBGS': 'BGS',
        'MINBIO': 'BIO', 'MINBIOI': 'BIOI',
        'SECDSL': 'DSLS', 'SECNGS': 'NGSS',
        'MINBIODSL': 'BIODSLS', 'SECBIODSL': 'BIODSLS'}

oper_cost_B = oper_cost_B.replace({'Sector': di_f})
oper_cost_B.rename(columns={'Sector':'FUEL'}, inplace=True)
oper_cost_B.head(5)
```

```python
# FILTERING SECTORS

# dictionary to filter out the rate of use by tech
di_rou = t_names.append(sector_namesS, ignore_index=True)
di_rou.drop(di_rou.iloc[:, 2:5], axis=1, inplace=True)
#di_rou
rof = n_file_r.loc[n_file_r['NAME']== 'RateOfUseByTech']
rof = rof.loc[rof['TECHNOLOGY'].isin(np.array(di_rou['TECHNOLOGY']))]

#pass the data frame in the function
use_tech = rate_us_ts(rof)

use_tech = use_tech.replace({'TECHNOLOGY': di2})
use_tech.loc[use_tech['TECHNOLOGY'].str.contains('CKGr'), 'TECHNOLOGY'] = 'Cooking rural'
use_tech.loc[use_tech['TECHNOLOGY'].str.contains('CKG'), 'TECHNOLOGY'] = 'Cooking urban'
use_tech.loc[use_tech['TECHNOLOGY'].str.contains('AGR'), 'TECHNOLOGY'] = 'Agriculture'
use_tech.loc[use_tech['TECHNOLOGY'].str.contains('IND'), 'TECHNOLOGY'] = 'Industry'
use_tech.loc[use_tech['TECHNOLOGY'].str.contains('DEMFSH'), 'TECHNOLOGY'] = 'Fisheries'
use_tech.loc[use_tech['TECHNOLOGY'].str.contains('|'.join(di6)),'TECHNOLOGY'] = 'Other'
use_tech = use_tech.groupby(['YEAR', 'TECHNOLOGY', 'FUEL'], as_index=False).sum()
use_tech.drop(use_tech.iloc[:, 4:13], axis=1, inplace=True)
```

```python
# data frame to calculate the fraction of fuel, then the fraction of cost
aux = use_tech.groupby(['YEAR','FUEL'], as_index=False).sum()

aux2 = pd.merge(use_tech, aux[['YEAR', 'FUEL', 'VALUE']] , on=['YEAR','FUEL'])
aux2['fraction'] = aux2['VALUE_x']/aux2['VALUE_y']
aux2.head(5)
```

```python
#merge B with aux2 from use by tech
oper_cost_C = pd.merge(oper_cost_B, aux2[['YEAR', 'TECHNOLOGY','FUEL', 'fraction']], on=['YEAR', 'FUEL'])
oper_cost_C['OC'] = oper_cost_C['VALUE']*oper_cost_C['fraction']
oper_cost_C = oper_cost_C.drop(oper_cost_C.columns[[1, 2, 4]], axis=1)
oper_cost_C.rename(columns={'TECHNOLOGY':'Sector', 'OC':'VALUE'}, inplace=True)
oper_cost_C.head(5)
```

```python
#final operating cost
oper_cost_D = oper_cost_A.append(oper_cost_C, ignore_index=True)
```
